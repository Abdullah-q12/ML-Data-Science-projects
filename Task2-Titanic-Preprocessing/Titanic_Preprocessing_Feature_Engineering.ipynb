```python
import pandas as pd
import seaborn as sb

df = sb.load_dataset("titanic")
df.head()
```

```python
df.isnull().sum()
```

```python
# Fill missing values without inplace=True (recommended way)
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])



```

```python
df.isnull().sum()
```

```python
df.select_dtypes(include=['number']).columns
```

```python
df.select_dtypes(exclude=['number']).columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Label Encoding for binary or yes/no type columns
binary_cols = ['sex', 'adult_male', 'alive', 'alone']
for col in binary_cols:
    df[col] = le.fit_transform(df[col])

# One-Hot Encoding for non-binary categorical columns
df = pd.get_dummies(df, columns=['embarked', 'class', 'who', 'embark_town'], drop_first=True)
```

```python
df.columns
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
num_cols = ['age', 'fare', 'sibsp', 'parch']  # choose your numeric columns except 'survived'
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
import pandas as pd

# Create a small sample DataFrame (just for learning)
sample_data = {
    'name': [
        'Braund, Mr. Owen Harris',
        'Cumings, Mrs. John Bradley (Florence Briggs Thayer)',
        'Heikkinen, Miss. Laina',
        'Allen, Mr. William Henry',
        'Moran, Mr. James',
        'McCarthy, Dr. Thomas',
        'Reynolds, Rev. George',
        'Palsson, Master. Gosta Leonard'
    ],
    'sibsp': [1, 1, 0, 0, 0, 0, 0, 3],
    'parch': [0, 0, 0, 0, 0, 0, 0, 1]
}

df = pd.DataFrame(sample_data)
df
```

```python
df.columns
```

```python
df['Title'] = df['name'].str.extract(' ([A-Za-z]+).', expand=False)
df['Title']
```

```python
df['Title'] = df['Title'].replace(['Mlle', 'Ms'], 'Miss')
df['Title'] = df['Title'].replace(['Mme'], 'Mrs')
df['Title'] = df['Title'].replace(['Dr', 'Rev', 'Major', 'Col', 'Capt', 'Sir', 'Don', 'Jonkheer', 'Lady', 'Countess'], 'Rare')
```

```python
df['Title'].value_counts()
```

```python
df['FamilySize'] = df['sibsp'] + df['parch'] + 1
df[['name', 'sibsp', 'parch', 'FamilySize']]
```

```python
df['FamilySize'].mean()
```

```python
df.to_csv('titanic_preprocessed.csv', index=False)
```

# Titanic Dataset Preprocessing and Feature Engineering

## Missing Value Imputation
- Numerical columns: Filled missing values with median.
- Categorical columns: Filled missing values with mode.

## Encoding
- Applied One-Hot Encoding to categorical features such as 'embarked', 'who', 'embark_town'.

## Scaling
- Used StandardScaler and MinMaxScaler to normalize numerical columns.

## Feature Engineering
- Extracted 'Title' from the 'Name' column.
- Grouped similar/rare titles (Dr, Rev  Rare).
- Created new feature 'FamilySize' = SibSp + Parch + 1.

## Results
- Unique Titles: Mr, Rare, Mrs, Miss, Master
- Average FamilySize: 1.75
- Final dataset saved as `titanic_preprocessed.csv`
